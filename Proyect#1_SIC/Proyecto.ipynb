```python
CLASS_MAX_GP = GP_math['absences'].max()
print(CLASS_MAX_GP)
```

```python
CLASS_MAX_MS = MS_math['absences'].max()
print(CLASS_MAX_MS)
```

```python
##### ESTUDIANTES QUE APRUEBAN Y REPRUEBAN (MATH) #####

AsistMS = ((CLASS_MAX_MS - MS_math['absences'])/CLASS_MAX_MS) * 100
AsistGP = ((CLASS_MAX_GP - GP_math['absences'])/CLASS_MAX_GP) * 100
AsistT_Math = AsistGP.add(AsistMS, fill_value = 0)
G3T_Math = GP_math['G3'].add(MS_math['G3'], fill_value = 0)

print(AsistT_Math)

As1 = AsistT_Math.lt(80) 
G3a = G3T_Math.lt(10)
G3b = G3T_Math.lt(15)
i = 0
appr = []
ex   = []

while(i < 395): 
    if(As1[i] == True):
        appr.append(0)
        ex.append(0)
        i+=1
    elif(As1[i] == False) & (G3a[i] == True):
        appr.append(0)
        ex.append(0)
        i+=1
    elif (As1[i] == False) & (G3a[i] == False) & (G3b[i] == True):
        appr.append(1)
        ex.append(1)    
        i+=1     
    elif (As1[i] == False) & (G3b[i] == False):
        appr.append(1)
        ex.append(0)
        i+=1 
        
Dataframe_colegio = Dataframe_colegio.assign(APPROVED = appr, EXTRA = ex)
Dataframe_colegio.head(395)
```
